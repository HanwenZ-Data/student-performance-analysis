# Student Performance Analysis

This project analyzes student scores across multiple subjects.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

df = pd.DataFrame({
    "student": ["A","B","C","D","E","F","G","H"],
    "math": [88,92,79,85,90,70,95,82],
    "english": [76,85,80,88,92,75,89,78],
    "econ": [90,87,78,84,91,72,94,80]
})

df.to_csv("../data/students.csv", index=False)
```

```python
df.describe()
```

```python
df["average"] = df[["math", "english", "econ"]].mean(axis=1)
df
```

```python
df.sort_values("average", ascending=False)
```

```python
df.plot(x="student", y=["math", "english", "econ"], kind="bar")
plt.title("Scores by Subject")
plt.show()
```

```python
df.plot(x="student", y="average", kind="bar", color="orange")
plt.title("Average Score")
plt.show()
```

## Insights

- Student G has the highest overall performance.
- Student F has the lowest scores across subjects.
- Math scores show higher variance than English.
